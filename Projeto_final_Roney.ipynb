## Análise da Progressão Salarial por Nível de Experiência

Vamos investigar como o salário médio varia entre os diferentes níveis de experiência (Entry-level - EN, Mid-level - MI, Senior-level - SE, Executive-level - EX) no conjunto de dados completo. Em seguida, compararemos essa progressão para os cargos principais de 'Data Analyst', 'Data Scientist' e 'Data Engineer'.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Mapear os níveis de experiência para uma ordem lógica para visualização
experience_order = ['Entry-level', 'Mid-level', 'Senior-level', 'Executive-level']

# Calcular o salário médio por nível de experiência (geral)
salary_by_experience_overall = df.groupby('experience_level')['salary_in_usd'].mean().reindex(experience_order)

print("Salário médio por nível de experiência (geral):")
print(salary_by_experience_overall)
```

### Salário Médio por Nível de Experiência (Geral)

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=salary_by_experience_overall.index, y=salary_by_experience_overall.values, palette='viridis')
plt.title('Salário Médio em USD por Nível de Experiência (Geral)')
plt.xlabel('Nível de Experiência')
plt.ylabel('Salário Médio (USD)')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

### Salário Médio por Nível de Experiência nos Cargos Principais

Agora, vamos filtrar para os cargos principais e comparar a progressão salarial.

```python
# Filtrar o DataFrame para os cargos principais
main_job_titles = ['Data Analyst', 'Data Scientist', 'Data Engineer']
df_main_jobs = df[df['job_title'].isin(main_job_titles)]

# Calcular o salário médio por cargo e nível de experiência
salary_by_job_experience = df_main_jobs.groupby(['job_title', 'experience_level'])['salary_in_usd'].mean().unstack()

# Reindexar as colunas de nível de experiência para a ordem correta
salary_by_job_experience = salary_by_job_experience[experience_order]

print("Salário médio por cargo e nível de experiência:")
print(salary_by_job_experience)
```

### Visualização da Progressão Salarial por Cargo

```python
salary_by_job_experience.T.plot(kind='line', figsize=(12, 7), marker='o')
plt.title('Progressão Salarial por Nível de Experiência para Cargos Principais')
plt.xlabel('Nível de Experiência')
plt.ylabel('Salário Médio (USD)')
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(title='Cargo')
plt.tight_layout()
plt.show()
```

### Conclusão

A análise da progressão salarial por nível de experiência revela que, de maneira geral, há um aumento constante no salário médio à medida que o nível de experiência avança, sendo 'Executive-level' o que oferece os maiores salários. Esta tendência de crescimento salarial com a experiência é consistente em todo o dataset.

Ao focar nos cargos principais ('Data Analyst', 'Data Scientist', 'Data Engineer'), observamos que a progressão salarial é similar entre eles, com salários crescentes para níveis de experiência mais altos. No entanto, o **Data Scientist** geralmente apresenta um salário médio mais elevado em todos os níveis de experiência em comparação com os outros dois cargos, seguido por **Data Engineer** e, por fim, **Data Analyst**.

## Análise Comparativa de Salários: Trabalho Remoto vs. Presencial

Vamos investigar se cargos totalmente remotos pagam mais, menos ou igual aos presenciais, e se essa diferença muda quando comparamos por nível de experiência.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Filtrar para 'Remote' e 'On-site' para a comparação principal
df_remote_vs_onsite = df[df['work_models'].isin(['Remote', 'On-site'])]

# Calcular o salário médio por modalidade de trabalho
salary_by_work_model = df_remote_vs_onsite.groupby('work_models')['salary_in_usd'].mean().sort_values(ascending=False)

print("Salário médio por modalidade de trabalho:")
print(salary_by_work_model)
```

### Salário Médio por Modalidade de Trabalho (Geral)

```python
plt.figure(figsize=(8, 5))
sns.barplot(x=salary_by_work_model.index, y=salary_by_work_model.values, palette='coolwarm')
plt.title('Salário Médio em USD: Remoto vs. Presencial')
plt.xlabel('Modalidade de Trabalho')
plt.ylabel('Salário Médio (USD)')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

### Salário Médio por Modalidade de Trabalho e Nível de Experiência

Agora, vamos analisar essa comparação detalhadamente por nível de experiência.

```python
# Calcular o salário médio por modalidade de trabalho e nível de experiência
salary_by_model_experience = df_remote_vs_onsite.groupby(['work_models', 'experience_level'])['salary_in_usd'].mean().unstack()

# Reindexar as colunas de nível de experiência para a ordem correta (se 'experience_order' estiver definido)
if 'experience_order' in locals() or 'experience_order' in globals():
    salary_by_model_experience = salary_by_model_experience[experience_order]

print("Salário médio por modalidade de trabalho e nível de experiência:")
print(salary_by_model_experience)
```

### Visualização da Comparação por Nível de Experiência

```python
salary_by_model_experience.T.plot(kind='bar', figsize=(12, 7), rot=45, cmap='plasma')
plt.title('Salário Médio em USD por Modalidade de Trabalho e Nível de Experiência')
plt.xlabel('Nível de Experiência')
plt.ylabel('Salário Médio (USD)')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend(title='Modalidade de Trabalho')
plt.tight_layout()
plt.show()
```

### Conclusão

A análise revela que, em termos gerais, **cargos remotos tendem a ter um salário médio ligeiramente superior** aos cargos presenciais no conjunto de dados.

Quando a comparação é feita por nível de experiência, a tendência se mantém para a maioria dos níveis: o trabalho remoto oferece salários médios mais altos, especialmente em níveis de experiência 'Mid-level' e 'Senior-level'. Pode haver variações, mas a compensação para trabalho remoto parece ser competitiva ou superior em grande parte da amostra.

## Análise das Melhores Oportunidades por País de Residência e Modalidade de Trabalho

Nesta seção, vamos investigar quais países de residência do empregado oferecem os maiores salários médios e se existe uma correlação entre o país e a modalidade de trabalho (remoto vs. presencial).

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calcular o salário médio por país de residência do empregado
salary_by_country = df.groupby('employee_residence')['salary_in_usd'].mean().sort_values(ascending=False)

print("Salário médio por país de residência (Top 10):")
print(salary_by_country.head(10))
```

### Países com os Maiores Salários Médios

```python
plt.figure(figsize=(12, 7))
sns.barplot(x=salary_by_country.head(10).index, y=salary_by_country.head(10).values, palette='viridis')
plt.title('Top 10 Países por Salário Médio em USD')
plt.xlabel('País de Residência do Empregado')
plt.ylabel('Salário Médio (USD)')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

### Correlação entre País e Modalidade de Trabalho

Agora, vamos analisar a distribuição das modalidades de trabalho (remoto, presencial, híbrido) para os países com os maiores salários médios.

```python
# Selecionar os top 5 países com maiores salários para análise detalhada
top_countries = salary_by_country.head(5).index.tolist()
df_top_countries = df[df['employee_residence'].isin(top_countries)]

# Contar a ocorrência de cada modalidade de trabalho por país
work_model_distribution = df_top_countries.groupby(['employee_residence', 'work_models']).size().unstack(fill_value=0)

# Normalizar para obter proporções
work_model_distribution_prop = work_model_distribution.apply(lambda x: x / x.sum(), axis=1)

print("Distribuição das modalidades de trabalho nos Top 5 países (proporção):")
print(work_model_distribution_prop)
```

### Visualização da Distribuição das Modalidades de Trabalho

```python
work_model_distribution_prop.plot(kind='bar', stacked=True, figsize=(12, 7), cmap='coolwarm')
plt.title('Distribuição das Modalidades de Trabalho nos Top 5 Países por Salário Médio')
plt.xlabel('País de Residência do Empregado')
plt.ylabel('Proporção')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Modalidade de Trabalho', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

### Conclusão

Ao analisar as melhores oportunidades, identificamos os países com os maiores salários médios. Destacam-se países como **Israel**, **Qatar**, **Malaysia**, **Puerto Rico** e **United States** com os salários mais elevados, sendo Israel o país com o maior salário médio.

Em relação à correlação entre país e modalidade de trabalho, observamos padrões distintos:

*   **Israel** mostra um cenário de 100% de trabalho presencial entre os registros analisados.
*   **Qatar** e **Malaysia** apresentam 100% de trabalho remoto.
*   **Puerto Rico** tem uma predominância de trabalho remoto (80%), com uma parte menor em modelo híbrido (20%).
*   Nos **Estados Unidos**, a maior parte dos empregados trabalha presencialmente (aproximadamente 61.7%), seguido por trabalho remoto (aproximadamente 37.5%), e uma pequena parcela em modelo híbrido (aproximadamente 0.8%).

Isso sugere que não há uma correlação uniforme entre altos salários e um único modelo de trabalho. Enquanto alguns países de alto salário são majoritariamente remotos, outros são predominantemente presenciais ou têm uma mistura, indicando que a modalidade de trabalho pode ser um fator relevante na estrutura salarial, mas sua prevalência varia significativamente por região.

## Análise Comparativa entre Cargos

Nesta seção, vamos comparar os cargos de 'Data Analyst', 'Data Scientist', 'Data Engineer' e 'Machine Learning Engineer' em termos de salário médio, número de vagas e crescimento salarial ao longo dos anos.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Definir os cargos de interesse
comparison_job_titles = ['Data Analyst', 'Data Scientist', 'Data Engineer', 'Machine Learning Engineer']

# Filtrar o DataFrame para os cargos de interesse
df_comparison = df[df['job_title'].isin(comparison_job_titles)]

print(f"DataFrame filtrado para os cargos: {comparison_job_titles}")
```

### Salário Médio por Cargo

```python
# Calcular o salário médio por cargo
salary_by_job_title = df_comparison.groupby('job_title')['salary_in_usd'].mean().sort_values(ascending=False)

print("Salário médio por cargo:")
print(salary_by_job_title)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=salary_by_job_title.index, y=salary_by_job_title.values, palette='coolwarm')
plt.title('Salário Médio em USD por Cargo')
plt.xlabel('Cargo')
plt.ylabel('Salário Médio (USD)')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

### Crescimento Salarial ao Longo dos Anos por Cargo

```python
# Calcular o salário médio por cargo e ano
salary_growth_by_year = df_comparison.groupby(['work_year', 'job_title'])['salary_in_usd'].mean().unstack()

print("Salário médio por cargo ao longo dos anos:")
print(salary_growth_by_year)
```

### Conclusão

Nesta análise comparativa, investigamos os cargos de 'Data Analyst', 'Data Scientist', 'Data Engineer' e 'Machine Learning Engineer' sob três perspectivas: salário médio, número de vagas e crescimento salarial ao longo dos anos.

**Salário Médio:**

**Número de Vagas:**

**Crescimento Salarial ao Longo dos Anos:**
*   Todos os cargos demonstram uma tendência geral de **crescimento salarial ao longo dos anos**, com uma notável ascensão dos salários em 2023 e 2024. ## Análise da Evolução Temporal

Nesta seção, vamos examinar como os salários médios e as modalidades de trabalho (remoto, presencial, híbrido) evoluíram entre os anos de 2020 e 2024.

### Variação do Salário Médio ao Longo dos Anos

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calcular o salário médio por ano
average_salary_per_year = df.groupby('work_year')['salary_in_usd'].mean()

print("Salário médio por ano:")
print(average_salary_per_year)
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x=average_salary_per_year.index, y=average_salary_per_year.values, marker='o', palette='viridis')
plt.title('Salário Médio em USD ao Longo dos Anos (2020-2024)')
plt.xlabel('Ano')
plt.ylabel('Salário Médio (USD)')
plt.grid(True, linestyle='--', alpha=0.7)
plt.xticks(average_salary_per_year.index)
plt.tight_layout()
plt.show()
```

### Mudanças na Modalidade de Trabalho ao Longo dos Anos

```python
# Contar a ocorrência de cada modalidade de trabalho por ano
work_model_yearly = df.groupby(['work_year', 'work_models']).size().unstack(fill_value=0)

# Normalizar para obter proporções por ano
work_model_yearly_prop = work_model_yearly.apply(lambda x: x / x.sum(), axis=1)

print("Proporção de modalidades de trabalho por ano:")
print(work_model_yearly_prop)
```

```python
plt.figure(figsize=(12, 7))
work_model_yearly_prop.plot(kind='bar', stacked=True, ax=plt.gca(), cmap='coolwarm')
plt.title('Distribuição das Modalidades de Trabalho ao Longo dos Anos (2020-2024)')
plt.xlabel('Ano')
plt.ylabel('Proporção')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend(title='Modalidade de Trabalho', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

### Conclusão

A análise da evolução temporal do mercado de trabalho em dados entre 2020 e 2024 revelou tendências importantes tanto nos salários médios quanto nas modalidades de trabalho:

**Variação do Salário Médio:**
*   Os salários médios na área de dados demonstraram um **crescimento consistente e significativo** ao longo dos anos, com um aumento notável de 2020 para 2021, e continuando a subir substancialmente em 2022, 2023 e 2024. Isso indica uma valorização contínua dos profissionais da área.

**Mudanças na Modalidade de Trabalho:**
*   Houve uma **mudança marcante na preferência por modalidades de trabalho** durante o período. Em 2020 e 2021, o trabalho presencial ('On-site') era dominante. No entanto, a partir de 2022, observamos um **aumento expressivo na proporção de vagas remotas ('Remote')**, superando o trabalho presencial em 2023 e mantendo uma forte presença em 2024. O modelo híbrido ('Hybrid') manteve uma proporção menor, mas estável, ao longo dos anos.
*   Essa transição para o trabalho remoto sugere uma adaptação do mercado às novas dinâmicas de trabalho, possivelmente impulsionada por fatores como a pandemia e a globalização das oportunidades.

Em resumo, o mercado de trabalho em dados está em **expansão salarial e com uma forte tendência de flexibilização da modalidade de trabalho**, com o modelo remoto se consolidando como uma opção preferencial para muitos profissionais e empresas.

## Porte da Empresa e Salário

Nesta seção, vamos investigar se empresas grandes (L), médias (M) ou pequenas (S) pagam mais, e se a diferença salarial é maior em algum nível de experiência específico.

### Salário Médio por Porte da Empresa (Geral)

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Definir a ordem para o porte da empresa para visualização
company_size_order = ['Small', 'Medium', 'Large']

# Calcular o salário médio por porte da empresa
salary_by_company_size = df.groupby('company_size')['salary_in_usd'].mean().reindex(company_size_order)

print("Salário médio por porte da empresa:")
print(salary_by_company_size)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=salary_by_company_size.index, y=salary_by_company_size.values, palette='plasma')
plt.title('Salário Médio em USD por Porte da Empresa')
plt.xlabel('Porte da Empresa')
plt.ylabel('Salário Médio (USD)')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

### Salário Médio por Porte da Empresa e Nível de Experiência

```python
# Calcular o salário médio por porte da empresa e nível de experiência
salary_by_size_experience = df.groupby(['company_size', 'experience_level'])['salary_in_usd'].mean().unstack()

# Reindexar as colunas de nível de experiência para a ordem correta
# Assuming 'experience_order' is defined from previous analysis, otherwise define it again
experience_order = ['Entry-level', 'Mid-level', 'Senior-level', 'Executive-level']
salary_by_size_experience = salary_by_size_experience[experience_order]

# Reindexar as linhas (company_size) para a ordem correta
salary_by_size_experience = salary_by_size_experience.reindex(company_size_order)

print("Salário médio por porte da empresa e nível de experiência:")
print(salary_by_size_experience)
```

```python
salary_by_size_experience.T.plot(kind='bar', figsize=(12, 7), rot=45, cmap='viridis')
plt.title('Salário Médio em USD por Porte da Empresa e Nível de Experiência')
plt.xlabel('Nível de Experiência')
plt.ylabel('Salário Médio (USD)')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend(title='Porte da Empresa', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

A análise do porte da empresa em relação aos salários revelou os seguintes padrões:

**Salário Médio por Porte da Empresa (Geral):**
